#1. Difference between a Function and a Method in Python

Function → Defined using def or lamda. Works independently.

Method → A function that belongs to an object (called using dot .).

✅ Example:

# Function
def add(a, b):
    return a + b

print(add(3, 5))   # Function call

# Method
text = "hello"
print(text.upper())   # .upper() is a method of string objects

#2. Function Arguments and Parameters

Parameter → Variable defined inside the function definition.

Argument → Actual value passed when calling the function.

✅ Example:

def greet(name):    # 'name' is a parameter
    print("Hello", name)

greet("Sneha")      # "Sneha" is the argument

#3. Different Ways to Define and Call a Function

Normal function:

def square(x):
    return x * x
print(square(4))


Lambda function:

square = lambda x: x * x
print(square(4))


Built-in functions:

print(len("hello"))

#4. Purpose of return Statement

It sends a value back from a function to the caller.

✅ Example:

def add(a, b):
    return a + b

result = add(2, 3)
print(result)   # 5

#5. Iterators vs Iterables

Iterable → Objects you can loop over (lists, tuples, strings).

Iterator → Produces items from iterable one by one using next().

✅ Example:

nums = [1, 2, 3]         # Iterable
it = iter(nums)          # Iterator
print(next(it))  # 1
print(next(it))  # 2

#6. Generators in Python

A special type of iterator defined using yield.

Produces values lazily (on demand).

✅ Example:

def fib(n):
    a, b = 0, 1
    for _ in range(n):
        yield a
        a, b = b, a + b

for num in fib(5):
    print(num)


Output: 0 1 1 2 3

#7. Advantages of Generators

Memory-efficient (don’t store entire sequence).

Useful for large datasets or infinite sequences.

✅ Example:

def count_up():
    n = 1
    while True:
        yield n
        n += 1

c = count_up()
print(next(c))  # 1
print(next(c))  # 2

#8. Lambda Function

Small anonymous function written in one line.

Used for short, temporary operations.

✅ Example:

square = lambda x: x*x
print(square(5))  # 25

#9. Purpose of map()

Applies a function to each element of an iterable.

✅ Example:

nums = [1, 2, 3, 4]
squares = map(lambda x: x*x, nums)
print(list(squares))  # [1, 4, 9, 16]

#10. Difference between map(), reduce(), and filter()

map(func, iterable) → Applies function to every element.

filter(func, iterable) → Keeps elements where function returns True.

reduce(func, iterable) → Reduces all elements to a single value.

✅ Example:

from functools import reduce

nums = [1, 2, 3, 4]

print(list(map(lambda x: x*2, nums)))   # [2, 4, 6, 8]
print(list(filter(lambda x: x%2==0, nums)))  # [2, 4]
print(reduce(lambda a,b: a+b, nums))    # 10

#11. Internal Mechanism of reduce() for Sum

List: [47, 11, 42, 13]
Function: lambda x,y: x+y

Step by step:

47 + 11 = 58

58 + 42 = 100

100 + 13 = 113

✅ Final result = 113

I. Write a Python function that takes a list of numbers as input and returns the sum of all even numbers in
the list.


```python
def add_even_numbers(l):  #l is a parameter which takes in the provided value
  sum=0    #initial value for sum is taken as 0
  for val in l:
    if val%2==0:
      sum+=val
  return sum
l=[1,2,3,4,6,7,8]
print(add_even_numbers(l))
```

5. Create an iterator class in Python that generates the ribonacci sequence up to a specified number of
terms.

```python
def fibonnaci(n):
  a=0
  b=1
  for i in range(0,n):
    yield a
    a,b = b,a+b

n = int(input())
gen = fibonnaci(n)
```

```python
print(next(gen))
```

 6.Write a generator function in Python that yields the powers of 2 up to a given exponent

```python
def powers_of_2(n):
  for i in range(0,n+1):
    yield 2**i
n = int(input())
gen = powers_of_2(n)
```

```python
print(next(gen))
```

7. Implement a generator function that reads a file line by line and yields each line as a string


```python
def read_file_by_line(file_path:str):
  with open(file_path,"r",encoding="utf-8") as file:
    for line in file:
      yield line.rstrip("\n")
gen = read_file_by_line("")  #path of file to be read
```
